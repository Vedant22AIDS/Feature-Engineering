```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
df['total_bill'].head()
```

```python
scaler.fit(df[['total_bill']])
```

```python
scaler.transform(df[['total_bill']])
```
